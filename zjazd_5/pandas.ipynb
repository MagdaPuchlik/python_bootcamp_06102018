2. Znajdź 10 meczów z największą różnicą bramek.

```python
dt["nowa_kolumna"]=1 # tworzenie nowej kolumny
dt["concatenate_names"] = dt["home_team"]+dt["away_team"]
dt["nowa+kolumna"]=abs(dt["home_team_goal"]-dt["away_team_goal"])
df = dt.sort_values(["nowa+kolumna"],ascending=False)
df[:10]
```

3. Policz liczbę zdobytych i straconych bramek na sezon dla wybranej drużyny. Narysuj wykres obrazujący uzyskane wyniki.

```python
team='Legia Warszawa'
team_matches = df[(df.home_team==team)|(df.away_team==team)]
team_matches.reset.index(inplace=True)
team_matches
```

```python
def team_goals(x):
    team = 'Legia Warszawa'
    if x.home_team == team:
        return x.home_team_goal
    else:
        return x.away_team_goal
def opponents_goals(x):
    team = 'Legia Warszawa'
    if x.home_team == team:
        return x.away_team_goal
    else:
        return x.home_team_goal
```

```python
team_matches['team_goals']=team_matches.apply(lambda x:x.home_team_goal if x.home_team == team else x.away_team_goal, axis=1)
# inny sposób team_matches = team_matches.apply(team_goals,axis=1)
```

```python
team_matches['opponents_goals'] = team_matches.apply(opponents_goals,axis=1)
team_matches.head()
```

```python
goals_per_season = team_matches.groupby('season').agg({'team_goals':['sum'],'opponents_goals':['sum']})
goals_per_season
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
goals_per_season.plot(kind="bar",stacked=True)
```
